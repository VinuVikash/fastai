```python
from fastai.vision import *
from fastai.metrics import error_rate
```

```python
path='C:/Users/NoobiWanKenobi/Documents/AI/seefood/data'
```

```python
data = ImageDataBunch.from_folder(path,ds_tfms=get_transforms(),size=224,bs=8).normalize(imagenet_stats)
```

```python
data.show_batch(rows=4,fig_size=(8,8))
```

```python
data.classes
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(3)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(3, figsize=(10,10))
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2,max_lr=1e-2)
```
